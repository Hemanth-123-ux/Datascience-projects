```python
#sample dataset of seaborn
import seaborn
import matplotlib.pyplot as plt
df=seaborn.load_dataset('tips')
print(df.to_string())
seaborn.pairplot(df)
plt.show()

```

```python
import seaborn as sns
import matplotlib.pyplot as plt
df=seaborn.load_dataset('tips')
selected_vars=['total_bill','tip']
sns.pairplot(df,vars=selected_vars)
plt.show()
```

```python
import seaborn
import matplotlib.pyplot as plt
df=seaborn.load_dataset('tips')
#size of the color density
seaborn.pairplot(df,hue='size')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
df=seaborn.load_dataset('tips')
#different colors
sns.pairplot(df,hue="size",palette="husl")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
df=seaborn.load_dataset('tips')
#kernel density function
sns.pairplot(df,diag_kind='kde')
plt.show()
```

```python
#adjusting plot kind
sns.pairplot(df, kind='reg')
plt.show()
```

```python
#controlling the markers
sns.pairplot(df, hue='sex', markers=["o","s"])
plt.show()
```

```python
#advanced customization of servers
g=sns.pairplot(df,hue='day')
g.fig.suptitle("pairplot of tips Dataset",y=1.02)
g.set(xticks=[],yticks=[])#remove tick labels
plt.show()
```
